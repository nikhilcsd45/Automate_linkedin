```python
from selenium import webdriver                            ##@@##11111
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time 
from selenium.webdriver.common.action_chains import ActionChains 

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open LinkedIn

driver.get('https://www.linkedin.com/home')

# try:
#     user_email = WebDriverWait(driver, 5).until(
#         EC.presence_of_element_located((By.ID, 'session_key'))
#     )
#     user_email.send_keys("nikhil7618987598@gmail.com")

#     user_password = WebDriverWait(driver, 15).until(
#         EC.presence_of_element_located((By.ID, 'session_password'))
#     )
#     user_password.send_keys("6386930858@Nik45")

#     sign_in = WebDriverWait(driver, 15).until(
#         EC.element_to_be_clickable((By.XPATH, '//*[@id="main-content"]/section[1]/div/div/form/div[2]/button'))
#     )
#     sign_in.click()

sign_in = WebDriverWait(driver, 15).until(
        EC.element_to_be_clickable((By.XPATH, "//*[@class='nav__button-secondary btn-md btn-secondary-emphasis']"))
    )
driver.execute_script("arguments[0].click();", sign_in)

email_box = WebDriverWait(driver, 15).until(
    EC.visibility_of_element_located((By.XPATH, "//*[@class='form__input--floating mt-24'][1]/input"))
)
driver.execute_script("arguments[0].value='nikhil7618987598@gmail.com';", email_box)

pass_box = WebDriverWait(driver, 15).until(
    EC.visibility_of_element_located((By.XPATH, "//*[@class='form__input--floating mt-24'][2]/input"))
)
driver.execute_script("arguments[0].value='6386930858@Nik45';", pass_box)

sign_in = WebDriverWait(driver, 15).until(
    EC.element_to_be_clickable((By.XPATH,"//*[@class='login__form_action_container ']/button"))
)
driver.execute_script("arguments[0].click();", sign_in)




element_to_click = WebDriverWait(driver, 20).until(
EC.element_to_be_clickable((By.XPATH,"//*[@class='msg-overlay-bubble-header__control msg-overlay-bubble-header__control--new-convo-btn artdeco-button artdeco-button--circle artdeco-button--muted artdeco-button--1 artdeco-button--tertiary ember-view'][2]"))
)
element_to_click.click()

# Navigate to the Jobs page
jobs = WebDriverWait(driver, 15).until(
EC.presence_of_element_located((By.XPATH, '//*[@id="global-nav"]/div/nav/ul/li[3]'))
)
jobs.click()

while True:
    try:
        show_all = WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.XPATH, "(//span[@class='artdeco-button__text' and text()='Show all'])[2]"))
        )
        # If found, break the loop
        break
    except:
        # Scroll down by 1000 pixels
        driver.execute_script("window.scrollBy(0, 1000);")
show_all.click()
search_inplace= WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.XPATH, "//*[@class='jobs-search-box__input-icon']"))
        )
actions = ActionChains(driver)
actions.click(search_inplace)
actions.key_down(Keys.COMMAND).send_keys('a').key_up(Keys.COMMAND).send_keys(Keys.DELETE).perform()
time.sleep(3)
actions.send_keys("India").perform()

search = WebDriverWait(driver, 15).until(
EC.presence_of_element_located((By.XPATH, '//*[@id="global-nav-search"]/div/div[2]/button[1]'))
)
time.sleep(3)
search.click()

all_filter = WebDriverWait(driver, 15).until(
EC.element_to_be_clickable((By.XPATH,"//*[@class='relative mr2']/button"))
)
all_filter.click()
print('successfully clicked on all_filter')
time.sleep(5)


for i in range(11):
    if i!=3 and i!=8 and i!=7 and i!=6 and i!=9:
        filters = driver.find_elements(By.XPATH, f"//*[@class='search-reusables__secondary-filters-filter'][{i+1}]/fieldset/div/ul/li")
        for j in range(len(filters)):
            xpath=f"//*[@class='search-reusables__secondary-filters-filter'][{i+1}]/fieldset/div/ul/li[{j+1}]/label/p"
            filter= driver.find_element(By.XPATH,xpath)
            # time.sleep(5)
            label_text = filter.text.strip()
            print(i,j,label_text)
            if i==0:
                if 'Most recent' in label_text:
                    try:
                        #WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH,xpath)))
                        filter.click()
                        time.sleep(5)
                        print('Clicked on element containing "recent"')
                        break 
                    except StaleElementReferenceException:
                        print("Caught a StaleElementReferenceException. Retrying...")
                        continue  
            elif i==1:
                if 'Past 24 hours' in label_text:
                    try:
                        filter.click()
                        time.sleep(5)
                        print('Clicked on element containing "recent"')
                        break 
                    except StaleElementReferenceException:
                        print("Caught a StaleElementReferenceException. Retrying...")
                        continue  
            elif i==2:
                if 'Internship' in label_text:
                    try:
                        filter.click()
                        time.sleep(5)
                        print('Clicked on element containing "recent"')
                        break  
                    except StaleElementReferenceException:
                        print("Caught a StaleElementReferenceException. Retrying...")
                        continue 
            elif i==4:
                if 'Internship' in label_text:
                    try:
                        filter.click()
                        time.sleep(5)
                        print('Clicked on element containing "recent"')
                        break 
                    except StaleElementReferenceException:
                        print("Caught a StaleElementReferenceException. Retrying...")
                        continue 
            elif i==5:
                if 'On-site' in label_text:
                    try:
                        filter.click()
                        time.sleep(5)
                        print('Clicked on element containing "recent"')
                        break 
                    except StaleElementReferenceException:
                        print("Caught a StaleElementReferenceException. Retrying...")
                        continue
            elif i==10:
                    if any(title in label_text for title in ['Software Engineer', 'Cloud Engineer', 'Full Stack Engineer', 'Senior Software Engineer']):
                        try:
                            filter.click()
                            time.sleep(5)
                            print('Clicked on element containing "recent"')
                        except StaleElementReferenceException:
                            print("Caught a StaleElementReferenceException. Retrying...")
                            continue
easy_apply = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "(//*[@class='artdeco-toggle__text'])[1]")))
easy_apply.click() 

filter_results=driver.find_element(By.XPATH,"//*[@class='justify-flex-end display-flex mv3 mh2']/button[2]")
filter_results.click()
 
time.sleep(5) 

description=driver.find_element(By.XPATH,"(//*[@class='mt4'])[2]")
text=description.text
print(text)
```
